```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# This data ranks cards by their chance of winning a trick in Euchre
# 'Next' Suit is the suit which is the same color but no the same suit as the 'Trump'
# 'Green' Suit is any suit which is the opposite color of the 'Trump'
```

```python
# Data from Euchre Universe
cards = ['Right Bower', 'Left Bower', 
         'Ace of Trump', 'King of Trump', 'Queen of Trump', 'Ten of Trump','Nine of Trump',
         'Ace of Next','King of Next','Queen of Next','Ten of Next','Nine of Next',
         'Ace of Green','King of Green','Queen of Green','Jack of Green','Ten of Green','Nine of Green']
performance = [100,76,
               53,45,40,34,32,
               44,13,2,1,1,
               50,19,6,1,1,1]
colors = ['r','r',
          'y','y','y','y','y',
          'b','b','b','b','b',
          'g','g','g','g','g','g']
ticks = np.arange(len(cards))
```

```python
#Print plot and index
plt.bar(ticks, performance, align='center', color=colors, alpha=0.5)
plt.xticks(ticks)
plt.ylabel('Chance of Winning')
plt.title('Euchre Cards')
plt.show()

pd.Series(np.arange(len(cards)), index=cards)
```

```python
# Sorts by performance
performanceNew, cardsNew, colorsNew = (list(t) for t in zip(*sorted(zip(performance, cards, colors), reverse=True)))
```

```python
#Print new plot and index
plt.bar(ticks, performanceNew, align='center', color=colorsNew, alpha=0.5)
plt.xticks(ticks)
plt.ylabel('Chance of Winning')
plt.title('Euchre Cards')
plt.show()

pd.Series(performanceNew, index=cards)
```
